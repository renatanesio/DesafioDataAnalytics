##### Data acquired from Kaggle ML Course

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

Get quantitative and qualitative data separately

```python
quantitative = [f for f in train.columns if train.dtypes[f] != 'object']
quantitative.remove('SalePrice') # This is our target
quantitative.remove('Id') # This is not useful
quantitative = sorted(quantitative)
qualitative = [f for f in train.columns if train.dtypes[f] == 'object']
qualitative = sorted(qualitative)
```

```python
print("Qualitative data:\n", qualitative)
```

```python
print("Quantitative data:\n", quantitative)
```

```python
# Visualizing sale prices
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,1,1)
sns.distplot(train['SalePrice'])
fig.add_subplot(2,1,2)
sns.boxplot(train['SalePrice'])
plt.tight_layout()
```

```python
def boxplot(x, y, **kwargs):
    sns.boxplot(x=x, y=y)
    x=plt.xticks(rotation=90)
f = pd.melt(train, id_vars=['SalePrice'], value_vars=qualitative)
g = sns.FacetGrid(f, col="variable",  col_wrap=2, sharex=False, sharey=False, height=5)
g = g.map(boxplot, "value", "SalePrice")
```
