```python
quantitative = train.select_dtypes(include=[np.number])

quantitative.head(10)
```

```python
qualitative = train.select_dtypes(exclude=[np.number])

qualitative.head(10)
```

```python
# Visualizing sale prices
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,1,1)
sns.distplot(train['SalePrice'])
fig.add_subplot(2,1,2)
sns.boxplot(train['SalePrice'])
plt.tight_layout()
```

```python
def boxplot(x, y, **kwargs):
    sns.boxplot(x=x, y=y)
    x=plt.xticks(rotation=60)
f = pd.melt(train, id_vars=['SalePrice'], value_vars=qualitative)
g = sns.FacetGrid(f, col="variable",  col_wrap=2, sharex=False, sharey=False, height=5)
g = g.map(boxplot, "value", "SalePrice")
```

Exterior quality, pool quality and the neighbourhood seem to play an important role in the sales price.
